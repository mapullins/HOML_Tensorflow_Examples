```python
import numpy as np
import pandas as pd
import matplotlib as mlt
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

# Sequential Model

```python
# Get the data
housing = fetch_california_housing()

# Create train, validation, test sets
X_train_full, X_test, y_train_full, y_test = train_test_split(housing.data, housing.target)
X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full)

# Scale the features since Gradient Descent is sensitive to scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_valid = scaler.transform(X_valid)
X_test = scaler.transform(X_test)
```

```python
# Create the model structure
model = keras.models.Sequential([
    keras.layers.Dense(30, activation = 'relu', input_shape = X_train.shape[1:]),
    keras.layers.Dense(1)
])
```

```python
model.summary()
```

```python
model.compile(loss = 'mean_squared_error', optimizer = 'sgd')
```

```python
# Train the model
history = model.fit(X_train,
                    y_train,
                    epochs = 40,
                    validation_data = (X_valid, y_valid))
```

```python
# Graph of training and validation performance
pd.DataFrame(history.history).plot(figsize = (8,5))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```

```python
# Check performance on test data
model.evaluate(X_test, y_test)
```

```python
# Pretend some of the test data are new instances to see how predict works
X_new = X_test[:3]
y_pred = model.predict(X_new)
for ii in range(3):
    print(f'Actual: {y_test[ii]}\nPredicted: {y_pred[ii][0]}\n')
```

# Functional API

```python
# Build a Wide and Deep model using keras functional api
input_ = keras.layers.Input(shape = X_train.shape[1:])
hidden1 = keras.layers.Dense(30, activation = 'relu')(input_)
hidden2 = keras.layers.Dense(30, activation = 'relu')(hidden1)
concat = keras.layers.Concatenate()([input_, hidden2])
output = keras.layers.Dense(1)(concat)

model = keras.Model(inputs = [input_], outputs = [output])
```

```python
model.summary()
```

```python
keras.utils.plot_model(model)
```

```python
model.compile(loss = 'mean_squared_error', 
              optimizer = keras.optimizers.SGD(learning_rate = .003) )
```

```python
# Train the model
history = model.fit(X_train,
                    y_train,
                    epochs = 40,
                    validation_data = (X_valid, y_valid))
```

```python
# Graph of training and validation performance
pd.DataFrame(history.history).plot(figsize = (8,5))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```

```python
# Check performance on test data
model.evaluate(X_test, y_test)
```

```python
# Pretend some of the test data are new instances to see how predict works
X_new = X_test[:3]
y_pred = model.predict(X_new)
for ii in range(3):
    print(f'Actual: {y_test[ii]}\nPredicted: {y_pred[ii][0]}\n')
```

```python
# Multiple input example
input_A = keras.layers.Input(shape = [5], name = 'wide_input')
input_B = keras.layers.Input(shape = [6], name = 'deep_input')
hidden1 = keras.layers.Dense(30, activation = 'relu')(input_B)
hidden2 = keras.layers.Dense(30, activation = 'relu')(hidden1)
concat = keras.layers.concatenate([input_A, hidden2])
output = keras.layers.Dense(1)(concat)

model = keras.Model(inputs = [input_A, input_B], outputs = [output])
model.summary()
```

```python
keras.utils.plot_model(model)
```

```python
model.compile(loss = 'mean_squared_error', 
              optimizer = keras.optimizers.SGD(learning_rate = .003) )
```

```python
# Since there are multiple inputs, we need to split up the training, validation, and testing sets to reflect this
X_train_A, X_train_B = X_train[:, :5], X_train[:, 2:]
X_valid_A, X_valid_B = X_valid[:, :5], X_valid[:, 2:]
X_test_A, X_test_B = X_test[:, :5], X_test[:, 2:]
X_new_A, X_new_B = X_test_A[:3], X_test_B[:3]
```

```python
# Train the model
history = model.fit((X_train_A, X_train_B),
                    y_train,
                    epochs = 40,
                    validation_data = ((X_valid_A, X_valid_B), y_valid))
```

```python
# Graph of training and validation performance
pd.DataFrame(history.history).plot(figsize = (8,5))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```

```python
# Check performance on test data
model.evaluate((X_test_A, X_test_B), y_test)
```

```python
# Pretend some of the test data are new instances to see how predict works
y_pred = model.predict((X_new_A, X_new_B))
for ii in range(3):
    print(f'Actual: {y_test[ii]}\nPredicted: {y_pred[ii][0]}\n')
```

```python
# Multiple Output Example
input_A = keras.layers.Input(shape = [5], name = 'wide_input')
input_B = keras.layers.Input(shape = [6], name = 'deep_input')
hidden1 = keras.layers.Dense(30, activation = 'relu')(input_B)
hidden2 = keras.layers.Dense(30, activation = 'relu')(hidden1)
concat = keras.layers.concatenate([input_A, hidden2])
output = keras.layers.Dense(1, name = 'main_output')(concat)
aux_output = keras.layers.Dense(1, name = 'aux_output')(hidden2)

model = keras.Model(inputs = [input_A, input_B], outputs = [output, aux_output])
model.summary()
```

```python
keras.utils.plot_model(model)
```

```python
# Loss functions are provided for each output; loss_weights are weighted to indicate the main output is more important
# aux_output is being used for regularization
model.compile(loss = ['mse', 'mse'],
              loss_weights = [0.9, 0.1],
              optimizer = keras.optimizers.SGD(learning_rate = .003) )
```

```python
# Train the model
history = model.fit((X_train_A, X_train_B),
                    (y_train, y_train),
                    epochs = 40,
                    validation_data = ((X_valid_A, X_valid_B), (y_valid, y_valid)))
```

```python
# Graph of training and validation performance
pd.DataFrame(history.history).plot(figsize = (8,5))
plt.grid(True)
plt.gca().set_ylim(0,2)
plt.show()
```

```python
# Check performance on test data
model.evaluate((X_test_A, X_test_B), (y_test, y_test))
```

```python
# Pretend some of the test data are new instances to see how predict works
y_pred_main, y_pred_aux = model.predict((X_new_A, X_new_B))
for ii in range(3):
    print(f'Actual: {y_test[ii]}\nPredicted Main: {y_pred_main[ii][0]}\nPredicted Aux: {y_pred_aux[ii][0]}\n')
```

# Subclass API

```python
# Create the previous model using the Subclass API
class WideAndDeepModel(keras.Model):
    
    def __init__(self, units = 30, activation = 'relu', **kwargs):
        super().__init__(**kwargs)
        self.hidden1 = keras.layers.Dense(units, activation = activation)
        self.hidden2 = keras.layers.Dense(units, activation = activation)
        self.main_output = keras.layers.Dense(1)
        self.aux_output = keras.layers.Dense(1)
        
    def call(self, inputs):
        input_A, input_B = inputs
        hidden1 = self.hidden1(input_B)
        hidden2 = self.hidden2(hidden1)
        concat = keras.layers.concatenate([input_A, hidden2])
        main_output = self.main_output(concat)
        aux_output = self.aux_output(hidden2)
        return main_output, aux_output
    
model = WideAndDeepModel()
```

```python
model.compile(loss = ['mse', 'mse'],
              loss_weights = [0.9, 0.1],
              optimizer = keras.optimizers.SGD(learning_rate = .003) )
```

```python
# Train the model
history = model.fit((X_train_A, X_train_B),
                    (y_train, y_train),
                    epochs = 40,
                    validation_data = ((X_valid_A, X_valid_B), (y_valid, y_valid)))
```

```python
# Graph of training and validation performance
pd.DataFrame(history.history).plot(figsize = (8,5))
plt.grid(True)
plt.gca().set_ylim(0,2)
plt.show()
```

```python
# Check performance on test data
model.evaluate((X_test_A, X_test_B), (y_test, y_test))
```

```python
# Pretend some of the test data are new instances to see how predict works
y_pred_main, y_pred_aux = model.predict((X_new_A, X_new_B))
for ii in range(3):
    print(f'Actual: {y_test[ii]}\nPredicted Main: {y_pred_main[ii][0]}\nPredicted Aux: {y_pred_aux[ii][0]}\n')
```
